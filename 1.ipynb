```python
#Q1 Convert data table into data frame and display the description

import pandas as pd
file = pd.read_excel("C:\Data\Anmol IDAP Assignment 2\cricket.xlsx")
data = pd.DataFrame(file)
data
```

```python
#Q2 Display the length of the dataset. Display first 10 records of the dataset. Display last five records of the data set.

k = len(data)
print("Length of dataset", k)

# First 10
data.head(10)
# Last five
data.tail(5)
```

```python
#Q3 Display the name of the player/players who played maximum match, maximum innings, scored maximum runs.

a = data.iloc[data.Mat.argmax(), 1]
c = data.iloc[data.Inns.argmax(), 1]
b = data.iloc[data.Runs.argmax(), 1]

print("Max Matches;", a)
print("Max Innings;", b)
print("Max Runs;", c)
```

```python
#Q4 Display the name of the player/players who played minimum match, minimum innings, scored minimum runs( non zero).

am = data.iloc[data.Mat.argmin(), 1]
cm = data.iloc[data.Inns.argmin(), 1]
bm = data.iloc[data.Runs.argmin(), 1]

print("Min Matches;", am)
print("Min Innings;", bm)
print("Min Runs;", cm)
```

```python
#Q5 Display total number of players and their average performance, country-wise.

tp = data['Player'].count()
print("Total Players in the dataset;", tp)
data[['Player', 'Ave']]
```

```python
#Q6 Display the player’s details who scored highest run in least BF.
# Data File has been manually cleaned to remove + in BF attribute
a = 0
b = 0
for index in data.index:
    if (data['Runs'][index] / data['BF'][index]) > b:
        b = (data['Runs'][index] / data['BF'][index])
        a = index
data.loc[a]
```
